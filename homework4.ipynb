### 2.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 100)

p = lambda x_: x_**2 + 1
gi = lambda x_: (x_ - 2)*(x - 4)
L = lambda x_, l: p(x_) + l*gi(x_)

g = lambda l: -9*l**2/(1 + l) + 8*l + 1
```

```python
plt.figure(figsize=(8, 8))

# Plot the objective x^2 + 1 versus x:
plt.plot(x, p(x), label="$objective: p(x)=x^2 + 1$")

plt.plot(x, gi(x), label="$condition: g(x)=(x-2)(x-4)$")

# Show the feasible set:
plt.axvspan(2, 4, color='blue', alpha=0.2, label="$feasible set: gi(x) \leq 0$")

# Show the optimal point and value:
plt.plot(2, 5, 'o', label="$optimal: x*=2, p*=5$")

# Plot the Lagrangian L(x, ) versus x for a few positive values of :
plt.plot(x, L(x, 1), label="$L(x, 1)=2x^2-6x+9$")
plt.plot(x, L(x, 2), label="$L(x, 2)=3x^2-12x+17$")
plt.plot(x, L(x, 3), label="$L(x, 3)=4x^2-24x+25$")

# Derive and sketch the Lagrange dual function g:
plt.plot(x, g(x), label="$g(\lambda)=-\frac{9 \lambda ^2}{1 + \lambda} + 8 \lambda + 1$")

plt.legend(loc='upper left')
plt.xlabel("$x$")
plt.grid(True);
```

The lower bound property verification:

The minimum value of $L(x, \lambda)$ over $x$ is always $< p*$. It increases as $\lambda$ varies from 0 to 2, reaches its maximum at $\lambda = 2$ and decreases again as $\lambda$ increases above $2$. We have equality $p* = g(\lambda)$ for $\lambda = 2$.

$L=x^2+1+\lambda(x-2)(x-4)=x^2+1+\lambda x^2 -6\lambda x+8\lambda=(1+\lambda)x^2-6\lambda x + 8\lambda +1$ - Langrangian

$L'=2(1+\lambda)x -6\lambda =0 \Rightarrow$

$L''=2(1+\lambda) \geq 0 \Rightarrow x*$ is a minimum.

$x*=\frac{3\lambda}{1+\lambda}$

$g(\lambda)=L(x*,\lambda)=\frac{(1+\lambda)9\lambda^2}{(1+\lambda)^2}-6\lambda \frac{3\lambda}{1+\lambda} + 8\lambda + 1=-\frac{9 \lambda^2}{1+\lambda}+8\lambda+1$ - Lagrange dual function g
